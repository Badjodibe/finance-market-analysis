```python
import pandas as pd
import numpy as np
import yfinance as yf
import pyspark
import matplotlib.pyplot as plt
```

```python
symbole = "EURUSD=X"
start_dateTr = "2000-01-01"
end_dateTr = "2016-12-31"
start_dateTe = "2017-01-01"
end_dateTe = "2020-12-31"
start_eval = "2021-01-01"
end_eval = "2023-05-31"
```

```python
train =  yf.download(symbole, start= start_dateTr, end=end_dateTr, interval = "1h")
test = yf.download(symbole, start=start_dateTe, end=end_dateTe)
eval_ =  yf.download(symbole, start=start_eval, end=end_eval)
```

```python
train = train.reset_index()
test = test.reset_index()
eval_ = eval_.reset_index()
```

```python
train.to_csv("train.csv")
test.to_csv("test.csv")
eval_.to_csv("eval.csv")
```

```python
import matplotlib.pyplot as plt

# Exemple de données pour les colonnes A et B
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 15, 8, 12, 6]
}

# Créer le DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Initialiser la figure
plt.figure()

# Afficher tous les points sur le même graphique avec une pause de 1 seconde entre chaque point
for i in range(len(df)):
    plt.plot(df['A'][i], df['B'][i], 'bo') # 'bo' pour des points bleus
    plt.xlabel('Axe X')
    plt.ylabel('Axe Y')
    plt.title('Graphique avec pauses entre les points')
    plt.xticks(df['A']) # Pour afficher les valeurs de A sur l'axe X
    plt.yticks(df['B']) # Pour afficher les valeurs de B sur l'axe Y
    plt.grid(True)
    plt.draw()
    plt.pause(1) # Pause de 1 seconde

# Afficher le graphique final
plt.show()
```
